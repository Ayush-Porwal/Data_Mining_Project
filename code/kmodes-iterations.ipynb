```python
## Initialisation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import mode

#Calculating distance:
def distance(x,y):
    j=0
    i=0
    k=0
    dist=[]
    #print(y.iloc[0])
    for j in range(len(x)):
        k=0
        for i in range(len(y)):
            if x.iloc[j,i]!=y.iloc[i]:
                k+=1
            i+=1
        dist.append(k)
        j+=1
    return dist               


%matplotlib inline
np.random.seed(200)
df=pd.read_csv('../data/mergedData.csv')
df['Test_Type']=df['Test_Type'].str.lower()
df['Signal_Strength'] =df['Signal_Strength'].abs()
df['Signal_Strength'] = df['Signal_Strength']*(-1)
df=df.sample(frac=0.0002, random_state=1011)
df=df.reset_index(drop=True)
df=df.drop(columns=['Data_Speed','Signal_Strength','LSA','Month&Year'])
dataframe=df
test_type=pd.get_dummies(df['Test_Type'])
df=df.drop(columns=['Test_Type'],axis=1)
df=pd.concat([test_type,df],axis=1)
technology=pd.get_dummies(df['Technology'])
df=df.drop(columns=['Technology'],axis=1)
df=pd.concat([technology,df],axis=1)
operator=pd.get_dummies(df['Operator'])
df=df.drop(columns=['Operator'],axis=1)
df=pd.concat([operator,df],axis=1)
print(dataframe)
print(df)

#np.random.seed(200)
n=3
i=0
#Initializing centroids without replacement
centroids = df.sample(n=3)
for i in range(10):
    
    #print(centroids)  
    #print(centroids.iloc[0,:])
    a=distance(df,centroids.iloc[0])
    b=distance(df,centroids.iloc[1])
    c=distance(df,centroids.iloc[2])
    a=pd.DataFrame(a,columns=['a'])
    b=pd.DataFrame(b,columns=['b'])
    c=pd.DataFrame(c,columns=['c'])
    dist=pd.concat([a,b,c],axis=1)
    df1=pd.DataFrame(dist.idxmin(axis=1),columns=['clusters'])
    data=pd.concat([df,dist,df1],axis=1) 
    print('data:\n',data)



    rslt_df1 = data.loc[(data['clusters'] == 'a')]
    rslt_df2= data.loc[(data['clusters'] == 'b')]
    rslt_df3 = data.loc[(data['clusters'] == 'c')]
    rslt_df=pd.concat([rslt_df1.mode(),rslt_df2.mode(),rslt_df3.mode()],axis=0)
    #print('rslt_df:\n',rslt_df)
    centroids=[]
    centroids=rslt_df.drop(columns=['a','b','c','clusters'])
    print(centroids)
    i+=1

```

```python
DF=pd.concat([dataframe,data.clusters],axis=1)

print(DF)
```
