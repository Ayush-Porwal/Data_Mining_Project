> # Reading the sampled csv dataset into numpy array

```python
import csv
with open('../data/associationSampleData.csv') as f:
    aprioriArray = list(csv.reader(f, delimiter=";"))
import numpy as np
aprioriArray = np.array(aprioriArray[1:])
```

```python
aprioriArray
```

> # using mlxtend library for association analysis

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_ary = te.fit(aprioriArray).transform(aprioriArray)
te_ary
```

```python
oht = TransactionEncoder()
oht_ary = oht.fit(aprioriArray).transform(aprioriArray)
df2 = pd.DataFrame(oht_ary, columns=oht.columns_)
print (df2)
```

> # Frequent itemsets from apriori with minimum support

```python
frequent_itemsets = apriori(df2, min_support=0.0005, use_colnames=True)
print (frequent_itemsets)
```

> # Possible association rules with min confidence

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets , metric="confidence", min_threshold=0)
```

> # dataframe for fpgrowth

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(aprioriArray).transform(aprioriArray)
df3 = pd.DataFrame(te_ary, columns=te.columns_)
df3
```

> # Frequent itemsets from fpgrowth with minimum support

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df3, min_support=0.0005)
```
