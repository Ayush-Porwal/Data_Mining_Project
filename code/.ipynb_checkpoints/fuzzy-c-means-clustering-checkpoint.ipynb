> # Fuzzy c Means Clustering

```python
import pandas as pd
import numpy as np
import operator
import math
import matplotlib.pyplot as plt
import random
from numpy import *
```

> ### Preprocessing of the data set for fuzzy c means:
> ##### [1] Only two columns Data speed and Signal strength are read
> ##### [2] All float objects are converted to int type after scaling normalized Data speed
> ##### [3] Random sampling of of the dataset is done

```python
data=pd.read_csv ('../data/mergedData.csv',usecols=['Data_Speed','Signal_Strength'])
columns = list(data.columns)
features = columns[:len(columns)]
data['Signal_Strength']=data.Signal_Strength.astype(int)
df = pd.DataFrame(data)
df["Data_Speed"] = 10000000000 * df["Data_Speed"]
data['Data_Speed']=data.Data_Speed.astype(int)
Sample=df.sample(frac=0.001)
Sample
```

> # passing the parameters for no.of clusters

```python
df = Sample[features]
attribute_count = len(df.columns)
k = 4 
MAX_ITER = 100
n = len(df)
m = 2.00
```

> # initializing membership matrix for randomly sampled data points 

```python
def initializeMembershipMatrix():
    membership_mat = list()
    for i in range(n):
        random_num_list = [random.random() for i in range(k)]
        summation = sum(random_num_list)
        temp_list = [x / summation for x in random_num_list]  #First normalization
        membership_mat.append(temp_list)
    return membership_mat
```

> # Calculating cluster centers from membership matrix

```python
def calculateClusterCenter(membership_mat):
    cluster_mem_val = zip(*membership_mat)
    cluster_centers = list()
    cluster_mem_val_list = list(cluster_mem_val)
    for j in range(k):
        x = cluster_mem_val_list[j]
        xraised = [e ** m for e in x]
        denominator = sum(xraised)
        temp_num = list()
        for i in range(n):
            data_point = list(df.iloc[i])
            prod = [xraised[i] * val for val in data_point]
            temp_num.append(prod)
        numerator = map(sum, zip(*temp_num))
        center = [z / denominator for z in numerator]  # Each dimension must be calculated.
        cluster_centers.append(center)
    return cluster_centers
```

> # Updating the values of membership matrix

```python
def updateMembershipValue(membership_mat, cluster_centers):
    #    p = float(2/(m-1))
    data = []
    for i in range(n):
        x = list(df.iloc[i])  # Take out each line of data in the file
        data.append(x)
        distances = [np.linalg.norm(list(map(operator.sub, x, cluster_centers[j]))) for j in range(k)]
        for j in range(k):
            den = sum([math.pow(float(distances[j] / distances[c]), 2) for c in range(k)])
            membership_mat[i][j] = float(1 / den)
    return membership_mat, data
```

> # Cluster labling the data points

```python
def getClusters(membership_mat):
    cluster_labels = list()
    for i in range(n):
        max_val, idx = max((val, idx) for (idx, val) in enumerate(membership_mat[i]))
        cluster_labels.append(idx)
    return cluster_labels
```

> # Defining fcmeans function with updated values through iterations

```python
def fuzzyCMeansClustering():
    # 
    membership_mat = initializeMembershipMatrix()
    curr = 0
    while curr <= MAX_ITER:  # The maximum number of iterations
        cluster_centers = calculateClusterCenter(membership_mat)
        membership_mat, data = updateMembershipValue(membership_mat, cluster_centers)
        cluster_labels = getClusters(membership_mat)
        curr += 1
    print(membership_mat)
    return cluster_labels, cluster_centers, data, membership_mat
```

> # Data point labels ; Cluster centers ; plot of clustered dataset with centres

```python
labels, centers, data, membership = fuzzyCMeansClustering()
print(labels)
```
