> ### Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

> ### Reading csv file into dataframe

```python
df=pd.read_csv('../data/mergedData.csv')
```

> ### Data_Speed VS Signal_Strength scatter plots for all operators

```python
jio=df.loc[df['Operator']=='JIO']
airtel=df.loc[df['Operator']=='AIRTEL']
idea=df.loc[df['Operator']=='IDEA']
vodafone=df.loc[df['Operator']=='VODAFONE']
cellone=df.loc[df['Operator']=='CELLONE']
uninor=df.loc[df['Operator']=='UNINOR']
aircel=df.loc[df['Operator']=='AIRCEL']
dolphin=df.loc[df['Operator']=='DOLPHIN']

jio.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('JIO')
plt.savefig('../plots/jio-scatter-plot.png', dpi=300, bbox_inches='tight')
airtel.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('AIRTEL')
plt.savefig('../plots/airtel-scatter-plot.png', dpi=300, bbox_inches='tight')
idea.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('IDEA')
plt.savefig('../plots/idea-scatter-plot.png', dpi=300, bbox_inches='tight')
vodafone.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('VODAFONE')
plt.savefig('../plots/vodafone-scatter-plot.png', dpi=300, bbox_inches='tight')
cellone.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('CELLONE')
plt.savefig('../plots/cellone-scatter-plot.png', dpi=300, bbox_inches='tight')
uninor.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('UNINOR')
plt.savefig('../plots/uninor-scatter-plot.png', dpi=300, bbox_inches='tight')
aircel.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('AIRCEL')
plt.savefig('../plots/aircel-scatter-plot.png', dpi=300, bbox_inches='tight')
dolphin.plot(kind='scatter',x='Data_Speed',y='Signal_Strength')
plt.title('DOLPHIN')
plt.savefig('../plots/dolphin-scatter-plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

> ### Normalised Average Data_Speed of Operators

```python
label=['JIO','AIRTEL','IDEA','VODAFONE','CELLONE','UNINOR','AIRCEL','DOLPHIN']
avg_data_speed=[]
for i in label:
    ser_provider=df[df['Operator']==i]
    avg=sum(ser_provider['Data_Speed'])/len(ser_provider['Data_Speed'])
    avg_data_speed.append(avg)

plt.figure(figsize=(10,6))
index=np.arange(len(label))
plt.bar(index, avg_data_speed)
plt.xlabel('Operator',fontsize=10)
plt.ylabel('Average-Data-Speed (normalised)',fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=45)
plt.title('Average_Data_Speed_of_Operators')
plt.savefig('../plots/Average_Data_Speed_of_Operators.png',dpi=300, bbox_inches='tight')
plt.show()
```

> ### Average Data_Speed in Licensed Service Areas

```python
avg_data_speed=[]
label=['Maharashtra','UP East','Andhra Pradesh','Madhya Pradesh','Gujarat','UP West','Tamil Nadu','Delhi','Kerala','Karnataka','Haryana','Rajasthan','Mumbai','Punjab','Bihar','Orissa','West Bengal','Kolkata','Jammu & Kashmir','Assam','Himachal Pradesh','North East','Chennai']
for i in label:
    state=df[df['LSA']==i]
    avg=sum(state['Data_Speed'])/len(state['Data_Speed'])
    avg_data_speed.append(avg)

plt.figure(figsize=(15,8))
index = np.arange(len(label))
plt.bar(index, avg_data_speed)
plt.xlabel('Licensed-Service-Area(LSA)', fontsize=15)
plt.ylabel('Average-Data-Speed(Normalised)', fontsize=10)
plt.xticks(index, label, fontsize=5, rotation=90)
plt.title('Average_Data_Speed_in_LSA')
plt.savefig('../plots/Average_Data_Speed_in_LSA.png', dpi=300, bbox_inches='tight')
plt.show()
```

> ### Pie chart plots for categorical features

```python
# pie chart of technology distribution in India
label = ['4G', '3G', '2G']
data = df['Technology'].value_counts()
plt.title('Technology_Distribution_In_India')
plt.pie(data, labels=label, autopct='%1.2f%%', shadow=True)
plt.savefig('../plots/pieplot-technology-distribution-in-India.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
# pie chart of technology distribution in India
label = ['upload', 'download']
data = df['Test_Type'].value_counts()
plt.title('Test_Type_Distribution_In_Data')
plt.pie(data, labels=label, autopct='%1.2f%%', shadow=True)
plt.savefig('../plots/test_type_distribution_in_data.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Operator distribution in India
label = ['JIO','AIRTEL','IDEA','VODAFONE','CELLONE','UNINOR','AIRCEL','DOLPHIN']
data = df['Operator'].value_counts()
plt.title('Operator_Distribution_In_Data')
plt.pie(data, labels=label, autopct='%1.2f%%', shadow=True)
plt.savefig('../plots/operator_distribution_in_data.png',dpi=300, bbox_inches='tight')
plt.show()
```
